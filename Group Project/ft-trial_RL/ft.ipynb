```python
from os import path
from pydub import AudioSegment
from scipy.fft import fft, fftfreq
import librosa
import matplotlib.pyplot as plt
import numpy as np
import time


def mp3_to_wav(audio):
    dst = audio.split(".")[0]+".wav"
    sound = AudioSegment.from_mp3(audio)
    sound.export(dst, format="wav")

    
def ft(audio):
    mp3_to_wav(audio)
    name = audio.split(".")[0] +".wav"
    x, data = librosa.load(name)
    N = len(x)
    c = fft(x) 
    d = fftfreq(N, 1 / data)  #
    plt.plot(d, np.abs(c))
    plt.show()
    
def main():
    ft("Vcut.mp3") #short
    ft("Viego.mp3") #long
    ft("Aimer-Stars in the rain.mp3") #friggin long

start_time = time.time()
main()
print("--- %s seconds ---" % (time.time() - start_time))

```
