## Question 1

The file `monthly-sunspots.txt` contains one floating point number per line. Write a program which opens the file, reads the numbers into a list, then plots the data as a line graph.


````python
import matplotlib.pyplot as plt
import numpy as np
name1 = "monthly-sunspots.txt"
def q1_file(file1): #returns list of data
    data = []
    with open(file1) as f:
        for flt_num in f:
            data.append(flt_num)
    return data
x = [i for i in range(len(q1_file(name1)))]
#print(len(q1_file(name1)))
plt.plot(x,q1_file(name1))
plt.show()
````

## Question 2

Write a program which writes a file `monthly-sunspots_cumulative.txt` which contains the cumulative total of the data in `monthly-sunspots.txt`. Make sure each number appears on a separate line.
```
58.0
120.6
190.6
...
```


````python
name2 = "monthly-sunspots_cumulative.txt"
with open(name2, "w") as f:
    cum_num = 0
    for i in range(len(q1_file(name1))):
            cum_num = cum_num + float(q1_file(name1)[i])
            f.write("{} \n".format(cum_num))
def cum_data(name):
    data2 = []
    with open(name) as f:
        for elem in f:
            data2.append(elem)
    return data2
#x2 = [i for i in range(len((name2)))]
#print(x2)
plt.plot(cum_data(name2))
plt.show()
````

## Question 4

Write a program which creates 9 files named `element_1.txt` to `element_9.txt` containing the name and description of the first 9 chemical elements listed in the file `periodic_table.csv`. E.g., `element_1.txt` should contain:
```
Hydrogen
colorless gas
```

````python
import csv
with open("periodic_table.csv") as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        num = int(row[0])
        name = row[1]
        appear = row[3]
        if num < 10:
            print(num, name, appear)
            with open("element_"+str(num)+".txt", "w") as k:
                k.write(name +"\n" + appear)
            k.close()
````
