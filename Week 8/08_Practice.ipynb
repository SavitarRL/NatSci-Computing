## Question 1

Create a 2 by 3 numpy array containing the following data:

$$
\begin{bmatrix}
10 & 12 & 13\\
5 & 6 & 7
\end{bmatrix}
$$

Use slice notation to create one-dimensional arrays containing:

1. The first row
1. The last column

Calculate:

1. A 3 element array containing the sum of each of the three columns
1. A 2 element array containing the sum of each of the two rows

```python
import numpy as np
arr = [[10,12,13],[5,6,7]]
x = np.array(arr)
print(x)

first_row = x[0,:]
print("First row 1D: {}".format(first_row))

last_col = x[:,-1]
print("Last column: {}".format(last_col))


sum_col = []
for i in range(len(x)+1):
    sum_col.append(sum(x[:,i]))

print("Sum of the 3 columns: {}".format(sum_col))

sum_row = []
for i in range(len(x)):
    sum_row.append(sum(x[i,:]))
print("Sum of the 2 rows: {}".format(sum_row))
```
