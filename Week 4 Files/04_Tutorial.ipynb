## Question 1

As a train accelerates out of the station, its position is measured as follows:

| time (s):    | 0  | 1 | 2 | 3 | 4  | 5  | 6 |
|--------------|----|---|---|---|---|---|---|
| position (m): |  0 |  0.15 | 0.61  | 1.35  | 2.40  | 3.75  | 5.41  |

1. Create two lists `time` and `speed` to store the data then plot a line graph showing the trains speed against time. Add a title `Train Speed` and axis labels `Time (s)` and `Speed (m/s)`.
1. Using the forumla $s =\frac{1}{2}at^2$, create a list `speed_est` containing the position of the train for each time point. Assume that $a=0.2~\mathrm{m/s^2}$.
1. Plot the `speed_est` on the graph. By varying $a$, estimate the acceleration of the train.


````python
# Answer
import matplotlib.pyplot as plt
position = [0,0.15,0.61,1.35,2.40,3.75,5.41]
time = [i for i in range(len(position))]
speed = [0, 0.15,0.61/2,1.35/3,2.4/4,3.75/5,5.41/6]

acceleration = [i*0.01 for i in range(1,21)]
#print(speed_est)
plt.figure(figsize=(10,10))
plt.plot(time, speed, 'k', label = "Real Speed")
for a in acceleration:
    speed_est = [a*t for t in time]
    plt.plot(time, speed_est, label = "Estimated speed: a = {} m/s^2".format(a))
plt.grid(True)
plt.xlabel("Time (s)") # add an x-axis label
plt.ylabel("Speed (m/s)") # add a y-axis label
plt.title("Train speed") # add a figure title
plt.legend(loc = "upper left")
plt.show()

#real 
````

## Question 4

A cannon located at position `x = 0` fires a cannonball at an angle `theta` from the horizontal (measured in radians) and speed `speed` m/s. Write a program which plots a graph of the cannonball's trajectory.

<img src="cannon.jpg" width="200px"></img>

Set the initial positions `x` and `y` to zero and velocities `vx` and `vy` to `np.cos(theta)` and `np.sin(theta)`. Each time step, update the positions using `x = x + vx * DELTA_T` and `y = y + vy * DELTA_T`. Update the velocity `vy = vy - g * DELTA_T`. The x velocity remains constant.
Create lists `x_list` and `y_list` to store the x and y positions, appending the new positions at each timestep.

Repeat until the `y` position becomes negative (what kind of loop should you use for this?). Finally, use `matplotlib` to plot the two lists.


Extend your program so that it prints `hit` or `miss` depending if the cannonball lands less than 20m from a target placed at a distance `dis` from the cannon.

<img src="cannon_output.png" width="200px"></img>

`hit`

````python
import numpy as np
import matplotlib.pyplot as plt

s = 93 # launch speed 93 m/s
theta = np.pi/4 # pi/4 radians equals 45 degrees

DELTA_T = 0.01
g = 9.8

x = 0 # initial x, y positions
y = 0
vx = s * np.cos(theta) # initial x, y velocities
vy = s * np.sin(theta)

dis = 900

x_pos = [0]
y_pos = [0]

x += vx*DELTA_T
y += vy*DELTA_T
vy -+ g*DELTA_T
x_pos.append(x)
y_pos.append(y)

while y > 0:
    x += vx*DELTA_T
    y += vy*DELTA_T
    vy -+ g*DELTA_T
    x_pos.append(x)
    y_pos.append(y)
plt.figure(figsize=(6,6))
plt.plot(x_pos, y_pos, "r-")
plt.xlabel("(m)")
plt.ylabel("(m)")
plt.title("Cannonball tradjectory")

if abs(dis-x) <= 20:
    print("Hit!")
````
