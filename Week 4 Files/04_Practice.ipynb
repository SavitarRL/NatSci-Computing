## Question 1
Use a loop to generate the following lists:
1. `[5, 7, 9, 11, 13, 15]`
1. `[1, 10, 100, 1000, 10000, 100000]`
1. `[0, 1, 2, 0, 1, 2, 0, 1, 2]` (hint: use the `%` operator)

## Question 2

Look up the predicted daily maximum and minimum temperatures for the next 10 days according to the BBC weather forecast.

https://www.bbc.co.uk/weather/2643743

Create two lists to store the maximum and minimum temperatures and plot them as two separate lines on a line graph, including axis labels and title. Look up how to add a legend.

<img src="temperatures.png" width="400px"/>

````python
import matplotlib.pyplot as plt
import numpy as np
high = [11, 9, 12, 12, 11, 10, 10, 11, 10, 10]
low = [6, 5, 10, 8, 7, 7,7,7,7,6]
day = range(0, len(high))

plt.figure(figsize=(5,5)) 

plt.plot(day,high , 'b', label = "Max")
plt.plot(day, low, 'g', label = "Min")
plt.grid(True)
plt.xlabel("Day") # add an x-axis label
plt.ylabel("Temperature C") # add a y-axis label
plt.title("Daily Temperature in Norfolk") # add a figure title
plt.legend(loc = "upper right")
plt.show()
````

## Question 3
Given a list of numbers `list_all`, create two lists `list_5` and `list_others`. `list_5` should contain all numbers divisible by 5, `list_others` should contain all other numbers.

```
list_all = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]

# your code here

print(list_5)
print(list_others)
```
Should output:
```
[15, 55, 75, 150, 180, 200]
[12, 32, 42, 122, 132]
```

````python
list_all = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]
list_5 = [i for i in list_all if i%5==0]
print(list_5)
list_others = [j for j in list_all if j%5!=0]
print(list_others)
````

## Question 4
The following code initialises a list `x` with ten random integers, then prints `x`.

Write code which prints four lines of output, containing
- Every element of `x` at an even index.
- Every even element.
- All elements in reverse order.
- Only the first and last element.

````python
import numpy as np
x = list(np.random.randint(20, size=10))
print("x =", x)
````

````python
print([x[i] for i in range(len(x)) if i%2 ==0])
print([i for i in x if i%2==0])
print(sorted(x, reverse=True))
print([x[i] for i in range(len(x)) if i==0 or i ==len(x)-1])
````
